### CSV 파일 통합

```python
import os
import pandas as pd
import glob

# 기본 경로 설정
base_path = r'C:\Users\d\OneDrive - 충북대학교\바탕 화면\RTP\어디쉐어'

# 하위 폴더 목록을 가져오기
folders = [f.path for f in os.scandir(base_path) if f.is_dir()]

# 각 폴더 내의 CSV 파일들을 하나로 합쳐서 저장
for folder in folders:
    # 폴더 내의 모든 CSV 파일 경로 검색
    csv_files = glob.glob(os.path.join(folder, '*.csv'))
    
    # 데이터프레임 리스트 초기화
    dataframes = []
    
    # 각 파일을 읽어서 데이터프레임으로 변환하고 리스트에 추가
    for file in csv_files:
        df = pd.read_csv(file)
        dataframes.append(df)
    
    # 모든 데이터프레임을 하나로 합치기
    if dataframes:  # 폴더에 CSV 파일이 있는 경우에만
        combined_df = pd.concat(dataframes, ignore_index=True)
        
        # 합쳐진 데이터프레임 저장
        combined_csv_path = os.path.join(folder, 'combined.csv')
        combined_df.to_csv(combined_csv_path, index=False)
        print(f'폴더 {folder}에 combined.csv 파일이 저장되었습니다.')
    else:
        print(f'폴더 {folder}에는 CSV 파일이 없습니다.')
```

### Isolation Forest 적용 - 어디쉐어(___ecfd1086a6934ae08b555b3ae880d31e)

```python
# 모듈 import 
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 데이터 불러오기
data = pd.read_csv(r'C:/Users/d/OneDrive - 충북대학교/바탕 화면/RTP/어디쉐어/___ecfd1086a6934ae08b555b3ae880d31e/combined.csv')
data = data.dropna(subset=['lat', 'lng'])

# 데이터 내 특정 열 지정
X = data[['lat', 'lng']]

X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# Isolation Forest 적용
iso_forest = IsolationForest(contamination=0.1, random_state=42)

iso_forest.fit(X_train)

predictions = iso_forest.predict(X_test)

X_test['anomaly'] = predictions

# 시각화
plt.figure(figsize=(10, 6))
plt.scatter(X_test[X_test['anomaly'] == 1]['lat'], X_test[X_test['anomaly'] == 1]['lng'], color='blue', s=5, label='Normal')
plt.scatter(X_test[X_test['anomaly'] == -1]['lat'], X_test[X_test['anomaly'] == -1]['lng'], color='red', s=5, label='Anomaly')
plt.title('Isolation Forest: Anomaly Detection')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.legend()
plt.show()
```

### 3차원 방법 확인

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler

# 데이터 불러오기 및 결측값 제거
data = pd.read_csv(r'C:/Users/d/OneDrive - 충북대학교/바탕 화면/RTP/어디쉐어/__d3b074a4d7b14c3db41661ee800e72b4/combined.csv')
data = data.dropna(subset=['lat', 'lng', 'time'])

# time 열을 datetime 형식으로 변환
data['time'] = pd.to_datetime(data['time'])

# time 열을 숫자 형식으로 변환 (초 단위로 변환)
data['time_numeric'] = data['time'].apply(lambda x: x.timestamp())

# lat, lng, time_numeric 열만 사용
X = data[['lat', 'lng', 'time_numeric']]

# train-test split
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# Isolation Forest 모델 학습
iso_forest = IsolationForest(contamination="auto", random_state=42)
iso_forest.fit(X_train)

# 예측
predictions = iso_forest.predict(X_test)

# 예측 결과를 데이터프레임에 추가
X_test['anomaly'] = predictions

# 3D 시각화
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# 정상 데이터 포인트
ax.scatter(X_test[X_test['anomaly'] == 1]['lat'], 
           X_test[X_test['anomaly'] == 1]['lng'], 
           X_test[X_test['anomaly'] == 1]['time_numeric'], 
           color='blue', s=5, label='Normal')

# 이상 데이터 포인트
ax.scatter(X_test[X_test['anomaly'] == -1]['lat'], 
           X_test[X_test['anomaly'] == -1]['lng'], 
           X_test[X_test['anomaly'] == -1]['time_numeric'], 
           color='red', s=5, label='Anomaly')

# 그래프 제목 및 축 라벨
ax.set_title('Isolation Forest: Anomaly Detection')
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_zlabel('Time (numeric)')

# 범례 추가
ax.legend()

# 그래프 보여주기
plt.show()
```

### 이상치 위도 경도 출력

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 데이터 로드 및 전처리
data = pd.read_csv(r'C:/Users/d/OneDrive - 충북대학교/바탕 화면/RTP/어디쉐어/__d3b074a4d7b14c3db41661ee800e72b4/combined.csv')
data = data.dropna(subset=['lat', 'lng'])

# 특성 선택
X = data[['lat', 'lng']]

# 데이터 분할
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# Isolation Forest 모델 생성 및 학습
iso_forest = IsolationForest(contamination="auto", random_state=42)
iso_forest.fit(X_train)

# 예측 및 점수 계산
predictions = iso_forest.predict(X_test)
scores = iso_forest.score_samples(X_test)

# 예측 결과를 DataFrame으로 변환
X_test_df = X_test.copy()
X_test_df['anomaly'] = predictions
X_test_df['score'] = scores

# 이상치 결과 필터링
anomalies = X_test_df[X_test_df['anomaly'] == -1]

# 이상치 및 점수 출력
print("Anomalies detected:")
print(anomalies)

# 시각화
plt.figure(figsize=(10, 6))
plt.scatter(X_test_df[X_test_df['anomaly'] == 1]['lat'], X_test_df[X_test_df['anomaly'] == 1]['lng'], color='blue', s=5, label='Normal')
plt.scatter(anomalies['lat'], anomalies['lng'], color='red', s=5, label='Anomaly')
plt.title('Isolation Forest: Anomaly Detection')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.legend()
plt.show()
```
