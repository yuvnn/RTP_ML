```python
import numpy as np

# 위도와 경도의 평균과 표준편차를 정의합니다.
mean_latitude = 37.5665  # 예: 서울의 위도
mean_longitude = 126.9780  # 예: 서울의 경도
std_dev = 0.01  # 위도와 경도의 표준편차

# 100개의 음식점 위치 데이터를 리스트로 생성합니다.
latitude = np.random.normal(mean_latitude, std_dev, 100)
longitude = np.random.normal(mean_longitude, std_dev, 100)

# 위도와 경도를 묶어서 리스트 형식으로 만듭니다.
data = list(zip(latitude, longitude))

print(data[:5])  # 처음 5개의 데이터 포인트를 출력
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# DBSCAN 모델을 생성합니다.
dbscan = DBSCAN(eps=0.01, min_samples=5)

# 데이터를 피팅하고 클러스터 라벨을 얻습니다.
clusters = dbscan.fit_predict(data)
print('clusters',clusters)

# 결과를 시각화합니다.
# 리스트를 numpy 배열로 변환하여 시각화에 사용합니다.
data_np = np.array(data)
print('data_np',data_np)
plt.figure(figsize=(10, 8))
plt.scatter(data_np[:, 1], data_np[:, 0], c=clusters, cmap='rainbow', s=50)
plt.title('DBSCAN Clustering of Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
